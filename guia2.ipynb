```python
import matplotlib.pyplot as plt

import numpy as np

from scipy.stats import skew
```

```python

# Crear un array de datos
data = np.array([14, 12, 21, 28, 30, 63, 29, 63, 55, 19, 20])

# Calcular la varianza
varianza = np.var(data)
print(f"Varianza: {varianza}")

# Calcular la desviación estándar
desviacion_estandar = np.std(data)
print(f"Desviación Estándar: {desviacion_estandar}")

# Calcular el coeficiente de asimetría
coeficiente_asimetria = skew(data)
print(f"Coeficiente de Asimetría: {coeficiente_asimetria}")
```

```python
# Crear el boxplot
plt.boxplot(data)

# Añadir título y etiquetas
plt.title('Boxplot del Array de Datos')
plt.ylabel('Valores')

# Mostrar el gráfico
plt.show()
```
